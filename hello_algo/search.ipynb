## 二分查找（binary search）
*question:* 给定一个长度为$n$ 的数组 $nums$ ，元素按从小到大的顺序排列且不重复。请查找并返回元素 $target$ 在该数组中的索引。若数组不包含该元素，则返回-1


```python
def binary_search(nums:list[int],target:int) -> int:
    """## 二分查找

    ### Args:
        - `num (list[int])`: _description_
        - `target (int)`: _description_

    ### Returns:
        - `int`: _description_
    """
    #初始化i， j
    i,j = 0,len(nums)-1
    while i <= j:
        m= ((j+i)//2)
        print('m: ',m)
        if nums[m] < target:
            i= m+1
        elif nums[m]  > target:
            j=m-1
        else:
            return m
    return -1

"""Driver Code"""
if __name__ == "__main__":
    target = 6
    nums = [1, 3, 6, 8, 12, 15, 23, 26, 31, 35]

    # 二分查找（双闭区间）
    index = binary_search(nums, target)
    print("目标元素 6 的索引 = ", index)
```

### 二分查找插入点

#### 无重复元素







```python
def binary_search_insertion_simple(nums: list[int], target: int) -> int:
    """二分查找插入点（无重复元素）"""
    i, j = 0, len(nums) - 1  # 初始化双闭区间 [0, n-1]
    while i <= j:
        m = (i + j) // 2  # 计算中点索引 m
        if nums[m] < target:
            i = m + 1  # target 在区间 [m+1, j] 中
        elif nums[m] > target:
            j = m - 1  # target 在区间 [i, m-1] 中
        else:
            return m  # 找到 target ，返回插入点 m
    # 未找到 target ，返回插入点 i
    return i


"""Driver Code"""
if __name__ == "__main__":
    # 无重复元素的数组
    nums = [1, 3, 6, 8, 12, 15, 23, 26, 31, 35]
    # 二分查找插入点
    target = 6
    index = binary_search_insertion_simple(nums, target)
    print(f"元素 {target} 的插入点的索引为 {index}")
    
    target = 11
    index = binary_search_insertion_simple(nums, target)
    print(f"元素 {target} 的插入点的索引为 {index}")
```

#### 有重复元素



```python
def binary_search_insertion(nums: list[int], target: int) -> int:
    """二分查找插入点（存在重复元素）"""
    i, j = 0, len(nums) - 1  # 初始化双闭区间 [0, n-1]
    while i <= j:
        m = (i + j) // 2  # 计算中点索引 m
        if nums[m] < target:
            i = m + 1  # target 在区间 [m+1, j] 中
        elif nums[m] > target:
            j = m - 1  # target 在区间 [i, m-1] 中
        else:
            j = m - 1  # 首个小于 target 的元素在区间 [i, m-1] 中
    # 返回插入点 i
    return i


"""Driver Code"""
if __name__ == "__main__":
    # 包含重复元素的数组
    nums = [1, 3, 6, 6, 6, 6, 6, 10, 12, 15]
    # 二分查找插入点
    target = 6
    index = binary_search_insertion(nums, target)
    print(f"元素 {target} 的插入点的索引为 {index}")
```

### 哈希优化策略



```python
# 解法一 暴力枚举
def two_sum_brute_force(nums: list[int], target: int) -> list[int]:
    """## 遍历所有

    ### Args:
        - `nums (list[int])`: _description_
        - `target (int)`: _description_

    ### Returns:
        - `list[int]`: _description_
    """
    for i in range(len(nums)-1):
        for j in range(i+1,len(nums)):
            if target == (nums[i]+nums[j]):
                return [i,j]
    return []

"""Driver Code"""
if __name__ == "__main__":
    nums = [2, 7, 11, 15]
    target = 13
    res = two_sum_brute_force(nums, target)
    print(res)
```

```python
def two_sum_hash_table(nums: list[int], target: int) -> list[int]:
    """方法二：辅助哈希表"""
    # 辅助哈希表，空间复杂度为 O(n)
    dic = {}
    # 单层循环，时间复杂度为 O(n)
    for i in range(len(nums)):
        if target - nums[i] in dic:
            return [dic[target - nums[i]], i]
        dic[nums[i]] = i
    return []

"""Driver Code"""
if __name__ == "__main__":
    nums = [2, 7, 11, 15]
    target = 26
    res = two_sum_hash_table(nums, target)
    print(res)
```

```python
def three_sum_hash_table(nums: list[int]):
    # 辅助哈希表，空间复杂度为 O(n)
    result = []
    # 单层循环，时间复杂度为 O(n)
    for i in range(len(nums)-1):
        dic = {}
        target = nums[i]
        for j in range(i+1,len(nums)):
            if (-1*target - nums[j]) in dic:
                result.append([target,(-1*target - nums[j]), nums[j]])
            dic[nums[j]] = j

    return result


nums = [2, 7, -9, 15,2,7]
print(nums.sort())
res = three_sum_hash_table(nums)
print(res)
```
