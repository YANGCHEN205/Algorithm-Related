*question:* $n$  $nums$  $target$ -1


```python
def binary_search(nums:list[int],target:int) -> int:
    """## 

    ### Args:
        - `num (list[int])`: _description_
        - `target (int)`: _description_

    ### Returns:
        - `int`: _description_
    """
    #i j
    i,j = 0,len(nums)-1
    while i <= j:
        m= ((j+i)//2)
        print('m: ',m)
        if nums[m] < target:
            i= m+1
        elif nums[m]  > target:
            j=m-1
        else:
            return m
    return -1

"""Driver Code"""
if __name__ == "__main__":
    target = 6
    nums = [1, 3, 6, 8, 12, 15, 23, 26, 31, 35]

    # 
    index = binary_search(nums, target)
    print(" 6  = ", index)
```

```python
def binary_search_insertion_simple(nums: list[int], target: int) -> int:
    """"""
    i, j = 0, len(nums) - 1  #  [0, n-1]
    while i <= j:
        m = (i + j) // 2  #  m
        if nums[m] < target:
            i = m + 1  # target  [m+1, j] 
        elif nums[m] > target:
            j = m - 1  # target  [i, m-1] 
        else:
            return m  #  target  m
    #  target  i
    return i


"""Driver Code"""
if __name__ == "__main__":
    # 
    nums = [1, 3, 6, 8, 12, 15, 23, 26, 31, 35]
    # 
    target = 6
    index = binary_search_insertion_simple(nums, target)
    print(f" {target}  {index}")
    
    target = 11
    index = binary_search_insertion_simple(nums, target)
    print(f" {target}  {index}")
```

```python
def binary_search_insertion(nums: list[int], target: int) -> int:
    """"""
    i, j = 0, len(nums) - 1  #  [0, n-1]
    while i <= j:
        m = (i + j) // 2  #  m
        if nums[m] < target:
            i = m + 1  # target  [m+1, j] 
        elif nums[m] > target:
            j = m - 1  # target  [i, m-1] 
        else:
            j = m - 1  #  target  [i, m-1] 
    #  i
    return i


"""Driver Code"""
if __name__ == "__main__":
    # 
    nums = [1, 3, 6, 6, 6, 6, 6, 10, 12, 15]
    # 
    target = 6
    index = binary_search_insertion(nums, target)
    print(f" {target}  {index}")
```

```python
#  
def two_sum_brute_force(nums: list[int], target: int) -> list[int]:
    """## 

    ### Args:
        - `nums (list[int])`: _description_
        - `target (int)`: _description_

    ### Returns:
        - `list[int]`: _description_
    """
    for i in range(len(nums)-1):
        for j in range(i+1,len(nums)):
            if target == (nums[i]+nums[j]):
                return [i,j]
    return []

"""Driver Code"""
if __name__ == "__main__":
    nums = [2, 7, 11, 15]
    target = 13
    res = two_sum_brute_force(nums, target)
    print(res)
```

```python
def two_sum_hash_table(nums: list[int], target: int) -> list[int]:
    """"""
    #  O(n)
    dic = {}
    #  O(n)
    for i in range(len(nums)):
        if target - nums[i] in dic:
            return [dic[target - nums[i]], i]
        dic[nums[i]] = i
    return []

"""Driver Code"""
if __name__ == "__main__":
    nums = [2, 7, 11, 15]
    target = 26
    res = two_sum_hash_table(nums, target)
    print(res)
```

```python
def three_sum_hash_table(nums: list[int]):
    #  O(n)
    result = []
    #  O(n)
    for i in range(len(nums)-1):
        dic = {}
        target = nums[i]
        for j in range(i+1,len(nums)):
            if (-1*target - nums[j]) in dic:
                result.append([target,(-1*target - nums[j]), nums[j]])
            dic[nums[j]] = j

    return result


nums = [2, 7, -9, 15,2,7]
print(nums.sort())
res = three_sum_hash_table(nums)
print(res)
```
