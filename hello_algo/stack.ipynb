```python
# 动态数组栈
class ArrayStack:
    """## 动态数组栈
    """
    def __init__(self) -> None:
        self._stack: list[int]=[]
    def size(self)->int:
        """## 链表长度

        ### Returns:
            - `int`: _description_
        """
        return len(self._stack)
    def is_empty(self)->bool:
        """## 链表是否为空

        ### Returns:
            - `bool`: _description_
        """
        return (self.size()==0)
    def push(self,val):
        """## 入栈

        ### Args:
            - `val (_type_)`: _description_
        """
        self._stack.append(val)
        
    def pop(self)->int:
        """## 出栈

        ### Returns:
            - `int`: _description_
        """
        if self.is_empty():
            raise IndentationError('empty Stack')
        return self._stack.pop()

    def peek(self) -> int:
        """访问栈顶元素"""
        if self.is_empty():
            raise IndexError("栈为空")
        return self._stack[-1]

    def to_list(self) -> list[int]:
        """返回列表用于打印"""
        return self._stack
```
