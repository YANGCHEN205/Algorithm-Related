```python
import collections
def levelOrder(self, root):
    """
    :type root: Node
    :rtype: List[List[int]]
    """
    if not root:
        return []
    queue = collections.deque([root])
    result = []
    while queue:
        level_value = []
        for i in range(len(queue)):
            node = queue.popleft()
            level_value.append(node.val)
            for child in node.children:
                queue.append(child)
        result.append(level_value)
    return result
```
