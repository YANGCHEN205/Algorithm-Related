给你一个满足下述两条属性的 m x n 整数矩阵：

每行中的整数从左到右按非严格递增顺序排列。
每行的第一个整数大于前一行的最后一个整数。
给你一个整数 target ，如果 target 在矩阵中，返回 true ；否则，返回 false 。

 

示例 1：



输入：matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 3
输出：true
示例 2：



输入：matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 13
输出：false
 


提示：

m == matrix.length
n == matrix[i].length
1 <= m, n <= 100
-104 <= matrix[i][j], target <= 104

```python
def binary_search(nums:list[int],target:int):
    left,right = 0,len(nums)-1
    while left <= right:
        mid = left + (right-left)//2
        if nums[mid]==target:
            return mid
        elif nums[mid] < target:
            left = mid+1
        else:
            right = mid-1
    return -1
    
nums=[1,2,3,4,5,6,7,87,89]
binary_search(nums,6)
```

```python
class Solution(object):
    
    def binary_search(self,nums:list[int],target:int):
        left,right = 0,len(nums)-1
        while left <= right:
            mid = left + (right-left)//2
            if nums[mid]==target:
                return mid
            elif nums[mid] < target:
                left = mid+1
            else:
                right = mid-1
        return left
    
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        if not matrix or not matrix[0]:
            return False

        rows, cols = len(matrix), len(matrix[0])
        
        # 对第一列进行二分查找以确定目标值可能所在的行
        top, bottom = 0, rows - 1
        while top <= bottom:
            mid_row = (top + bottom) // 2
            if matrix[mid_row][0] == target:
                return True
            elif matrix[mid_row][0] < target:
                top = mid_row + 1
            else:
                bottom = mid_row - 1

        # 如果退出循环时，bottom < 0，说明目标值小于矩阵最小值
        if bottom < 0:
            return False

        # target 应该在第 bottom 行（top 行与 bottom 行相交）
        row_to_search = bottom

        # 对选定的行进行二分查找
        left, right = 0, cols - 1
        while left <= right:
            mid_col = (left + right) // 2
            if matrix[row_to_search][mid_col] == target:
                return True
            elif matrix[row_to_search][mid_col] < target:
                left = mid_col + 1
            else:
                right = mid_col - 1

        return False

matrix = [
    [1, 3, 5, 7],
    [10, 11, 16, 20],
    [23, 30, 34, 60]
]
sol =Solution()
sol.searchMatrix(matrix,3)
```
