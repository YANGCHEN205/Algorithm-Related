```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isSameTree(self, p, q):
        """
        :type p: TreeNode
        :type q: TreeNode
        :rtype: bool
        """
        if not p and not q: return True
        
        return self.compare(p, q)
    
    def compare(self,p,q):
        #
        if p == None and q != None: return False
        elif p != None and q  == None: return False
        elif p == None and q == None: return True
        #
        elif p.val != q.val: return False
        
        #
        #
        leftside = self.compare(p.left, q.left) # 
        rightside = self.compare(p.right, q.right) # 
        isSame = leftside and rightside #  
        return isSame
    
```
