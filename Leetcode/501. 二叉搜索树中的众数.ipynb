501. 二叉搜索树中的众数
简单
相关标签
相关企业
给你一个含重复值的二叉搜索树（BST）的根节点 root ，找出并返回 BST 中的所有 众数（即，出现频率最高的元素）。

如果树中有不止一个众数，可以按 任意顺序 返回。

假定 BST 满足如下定义：

结点左子树中所含节点的值 小于等于 当前节点的值
结点右子树中所含节点的值 大于等于 当前节点的值
左子树和右子树都是二叉搜索树
 

示例 1：


输入：root = [1,null,2,2]
输出：[2]
示例 2：

输入：root = [0]
输出：[0]
 

提示：

树中节点的数目在范围 [1, 104] 内
-105 <= Node.val <= 105
 

进阶：你可以不使用额外的空间吗？（假设由递归产生的隐式调用栈的开销不被计算在内）

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def __init__(self):
        self.pre = None
        self.count = 0
        self.result = []
        self.maxcount = 0
    def searchBST(self,cur):
        if not cur:
            return
        self.searchBST(cur.left)
        if not self.pre:
            self.count = 1
        elif self.pre.val ==cur.val:
            self.count+=1
        else:
            self.count = 1
        self.pre = cur
        if self.count == self.maxcount:
            self.result.append(cur.val)
        if self.count > self.maxcount:
            self.maxcount = self.count
            self.result = []
            self.result.append(cur.val)
        self.searchBST(cur.right)
        return
    def findMode(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        self.searchBST(root)
        return self.result
```

```python
## 前中后序列递归实现
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def __init__(self):
        return
    def preorder_traversal(self,root):
        return [root.val]+self.preorder_traversal(root.left)+self.preorder_traversal(root.right) if root else []
    def inorder_traversal(self,root):
        return self.inorder_traversal(root.left) + [root.val] + self.inorder_traversal(root.right) if root else []
    def postorder_traversal(self,root):
        return self.postorder_traversal(root.left)+self.postorder_traversal(root.right)+[root.val] if root else []
# 测试数据
root = TreeNode(1, TreeNode(2), TreeNode(3, TreeNode(4), TreeNode(5)))
sol = Solution()
print("前序遍历:", sol.preorder_traversal(root))
print("中序遍历:", sol.inorder_traversal(root))
print("后序遍历:", sol.postorder_traversal(root))
```

```python
## 循环遍历所有元素
from collections import defaultdict
class Solution(object):
    def __init__(self):
        self.freq_map = defaultdict(int)
        return
    def searchBST(self,cur):
        if not cur:
            return 
        self.freq_map[cur.val] += 1
        self.searchBST(cur.left)
        self.searchBST(cur.right)
    def findMode(self,root):
        result = []
        if not root:
            return []
        self.searchBST(root)
        max_freq = max(self.freq_map.values())
        for key, freq in self.freq_map.items():
            if freq == max_freq:
                result.append(key)
        return result
        
```
