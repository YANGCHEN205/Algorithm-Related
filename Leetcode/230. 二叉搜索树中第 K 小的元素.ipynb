提示
给定一个二叉搜索树的根节点 root ，和一个整数 k ，请你设计一个算法查找其中第 k 小的元素（从 1 开始计数）。

 

示例 1：



输入：root = [3,1,4,null,2], k = 1
输出：1
示例 2：


输入：root = [5,3,6,2,4,null,null,1], k = 3
输出：3
 

 

提示：

树中的节点数为 n 。
1 <= k <= n <= 104
0 <= Node.val <= 104

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def __init__(self):
        self.k = 0
        self.result = 0
    def in_order_traversal(self,node):
        if node is None:
            return
        self.in_order_traversal(node.left)
        self.k -= 1
        if self.k == 0:
            self.result = node.val
            return
        self.in_order_traversal(node.right)
        return
    def kthSmallest(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: int
        """
        self.k = k
        self.in_order_traversal(root)
        return self.result
# 测试样例
# 构造一个二叉搜索树:
#       3
#      / \
#     1   4
#      \
#       2
root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(4)
root.left.right = TreeNode(2)
sol = Solution()
k = 3
print(sol.kthSmallest(root, k))  # 输出：1
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def kth_smallest(root: TreeNode, k: int) -> int:
    def in_order_traversal(node):
        if node is None:
            return []
        return in_order_traversal(node.left) + [node.val] + in_order_traversal(node.right)
    
    in_order = in_order_traversal(root)
    return in_order[k - 1]

# 测试样例
# 构造一个二叉搜索树:
#       3
#      / \
#     1   4
#      \
#       2
root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(4)
root.left.right = TreeNode(2)

k = 1
print(kth_smallest(root, k))  # 输出：1

k = 2
print(kth_smallest(root, k))  # 输出：2

k = 3
print(kth_smallest(root, k))  # 输出：3

k = 4
print(kth_smallest(root, k))  # 输出：4
```

解法 2：优化的中序遍历
在解法 1 中，我们进行了完整的中序遍历，这样时间复杂度是 𝑂(𝑛).实际上，我们只需要遍历到第𝑘小的元素就可以停止遍历，从而节省时间。

```python
def kth_smallest_optimized(root: TreeNode, k: int) -> int:
    def in_order_traversal(node):
        nonlocal k, result
        if node is None:
            return
        in_order_traversal(node.left)
        k -= 1
        if k == 0:
            result = node.val
            return
        in_order_traversal(node.right)
    
    result = None
    in_order_traversal(root)
    return result

# 测试样例与解法 1 相同
```

解法 4：基于栈的迭代中序遍历

```python
def kth_smallest_iterative(root: TreeNode, k: int) -> int:
    stack = []
    current = root
    
    while True:
        while current is not None:
            stack.append(current)
            current = current.left
        
        current = stack.pop()
        k -= 1
        if k == 0:
            return current.val
        
        current = current.right

# 测试样例与解法 1 相同
```
