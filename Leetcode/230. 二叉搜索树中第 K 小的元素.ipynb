```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def __init__(self):
        self.k = 0
        self.result = 0
    def in_order_traversal(self,node):
        if node is None:
            return
        self.in_order_traversal(node.left)
        self.k -= 1
        if self.k == 0:
            self.result = node.val
            return
        self.in_order_traversal(node.right)
        return
    def kthSmallest(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: int
        """
        self.k = k
        self.in_order_traversal(root)
        return self.result
# ÊµãËØïÊ†∑‰æã
# ÊûÑÈÄ†‰∏Ä‰∏™‰∫åÂèâÊêúÁ¥¢Ê†ë:
#       3
#      / \
#     1   4
#      \
#       2
root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(4)
root.left.right = TreeNode(2)
sol = Solution()
k = 3
print(sol.kthSmallest(root, k))  # ËæìÂá∫Ôºö1
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def kth_smallest(root: TreeNode, k: int) -> int:
    def in_order_traversal(node):
        if node is None:
            return []
        return in_order_traversal(node.left) + [node.val] + in_order_traversal(node.right)
    
    in_order = in_order_traversal(root)
    return in_order[k - 1]

# ÊµãËØïÊ†∑‰æã
# ÊûÑÈÄ†‰∏Ä‰∏™‰∫åÂèâÊêúÁ¥¢Ê†ë:
#       3
#      / \
#     1   4
#      \
#       2
root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(4)
root.left.right = TreeNode(2)

k = 1
print(kth_smallest(root, k))  # ËæìÂá∫Ôºö1

k = 2
print(kth_smallest(root, k))  # ËæìÂá∫Ôºö2

k = 3
print(kth_smallest(root, k))  # ËæìÂá∫Ôºö3

k = 4
print(kth_smallest(root, k))  # ËæìÂá∫Ôºö4
```

```python
def kth_smallest_optimized(root: TreeNode, k: int) -> int:
    def in_order_traversal(node):
        nonlocal k, result
        if node is None:
            return
        in_order_traversal(node.left)
        k -= 1
        if k == 0:
            result = node.val
            return
        in_order_traversal(node.right)
    
    result = None
    in_order_traversal(root)
    return result

# ÊµãËØïÊ†∑‰æã‰∏éËß£Ê≥ï 1 Áõ∏Âêå
```

```python
def kth_smallest_iterative(root: TreeNode, k: int) -> int:
    stack = []
    current = root
    
    while True:
        while current is not None:
            stack.append(current)
            current = current.left
        
        current = stack.pop()
        k -= 1
        if k == 0:
            return current.val
        
        current = current.right

# ÊµãËØïÊ†∑‰æã‰∏éËß£Ê≥ï 1 Áõ∏Âêå
```
