```python
def largestValues(self, root):
    if not root:
        return []
    queue = collections.deque([root])
    result = []
    while queue:
        layer_value = []
        len_queue = len(queue)
        for i in range(len_queue):
            node = queue.popleft()
            layer_value.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        result.append(max(layer_value))
    return result
```
