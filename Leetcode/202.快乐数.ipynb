# 202. 快乐数

编写一个算法来判断一个数 n 是不是快乐数。

「快乐数」 定义为：

对于一个正整数，每一次将该数替换为它每个位置上的数字的平方和。
然后重复这个过程直到这个数变为 1，也可能是 无限循环 但始终变不到 1。
如果这个过程 结果为 1，那么这个数就是快乐数。
如果 n 是 快乐数 就返回 true ；不是，则返回 false 。

 

示例 1：

输入：n = 19
输出：true
解释：
12 + 92 = 82
82 + 22 = 68
62 + 82 = 100
12 + 02 + 02 = 1
示例 2：

输入：n = 2
输出：false
 
 

```python
def isHappy(n):
    """
    :type n: int
    :rtype: bool
    """
    def get_next(n):
        total_sum = 0
        while n > 0:
            n, digit = divmod(n, 10)
            total_sum += digit ** 2
        return total_sum

    num_map =set()
    while n!=1 and n not in num_map:
        num_map.add(n)
        n = get_next(n)
    return n==1

isHappy(7)
```

```python
def isHappy(n):
    """
    :type n: int
    :rtype: bool
    """
    def get_next(n):
        total_sum = 0
        while n > 0:
            n, digit = divmod(n, 10)
            total_sum += digit ** 2
        return total_sum
    slower = n
    faster = get_next(n)
    while faster !=1 and slower != faster:
        slower=get_next(slower)
        faster=get_next(get_next(faster))
    return faster==1

isHappy(17)
```
