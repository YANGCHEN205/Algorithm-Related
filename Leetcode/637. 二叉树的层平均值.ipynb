```python

import collections
def averageOfLevels(root):
    """
    :type root: TreeNode
    :rtype: List[float]
    """
    if not root:
        return []
    queue = collections.deque([root])
    result = []
    while queue:
        layer_value = 0
        len_queue = len(queue)
        for i in range(len_queue):
            node = queue.popleft()
            layer_value += node.val
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        print(layer_value)
        result.append(layer_value/len_queue)
    return result
            
```
