提示
给定一个只包括 '('，')'，'{'，'}'，'['，']' 的字符串 s ，判断字符串是否有效。

有效字符串需满足：

左括号必须用相同类型的右括号闭合。
左括号必须以正确的顺序闭合。
每个右括号都有一个对应的相同类型的左括号。
 

示例 1：

输入：s = "()"
输出：true
示例 2：

输入：s = "()[]{}"
输出：true
示例 3：

输入：s = "(]"
输出：false
 

提示：

1 <= s.length <= 104
s 仅由括号 '()[]{}' 组成


```python
def isValid(s):
    """
    :type s: str
    :rtype: bool
    """
    str_stack = []
    mapping = {
            '(': ')',
            '[': ']',
            '{': '}'
        }
    for item in s:
        if item in mapping.keys():
            str_stack.append(mapping[item])
        elif not str_stack or str_stack[-1] != item:
            return False
        else:
            str_stack.pop()
    return True if not str_stack else False

s='[{}(])'
isValid(s)
```
