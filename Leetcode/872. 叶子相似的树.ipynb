```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def leafSimilar(self, root1, root2):
        """
        :type root1: TreeNode
        :type root2: TreeNode
        :rtype: bool
        """
        def dfs(root):
            if root.left is None and root.right is None:
                return f"{root.val}"
            leaf_seq = ""
            if root.left:
                leaf_seq+=dfs(root.left)
            if root.right:
                leaf_seq+=dfs(root.right)
            return leaf_seq
        leaf_seq1 = dfs(root1) if root1 else ""
        leaf_seq2 = dfs(root2) if root2 else ""
        return leaf_seq1==leaf_seq2
```
